```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, test_size=0.2)
```

```python
train.drop("Date", axis=1, inplace=True)
test.drop("Date", axis=1, inplace=True)
```

```python
train.head()
```

```python
# Removing punctuations
import re
train_data = train.iloc[:,1:27]
train_data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
y = train.iloc[:,0]
```

```python
y = pd.DataFrame(y)
```

```python
# Renaming column names for ease of access
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
train_data.columns = new_index
```

```python
train_data.head()
```

```python
#Converting all columns to lower case
for index in new_index:
    train_data[index] = train_data[index].str.lower()

train_data.head()
```

```python
" ".join(str(x) for x in train_data.iloc[1,0:25])
```

```python
# Merging all the columns to convert into one paragraph text column
headlines = []
for row in range(0, len(train_data)):
    headlines.append(" ".join(str(x) for x in train_data.iloc[row,0:25]))
```

```python
headlines[1]
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
stop = stopwords.words("english")
lem = WordNetLemmatizer()
```

```python
# Removing stop words, applying lemmitization
corpus = []
for i in range(len(headlines)):
    review = headlines[i]
    review = review.split()
    review = [lem.lemmatize(word) for word in review if word not in stop]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus[1]
```

```python
# Applying TF-IDF to corpus text to convert into numeric vector
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(ngram_range=(2,2))
```

```python
X = tfidf.fit_transform(corpus)
```

```python
X
```

```python
# Training the model on train Data
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest = RandomForestClassifier(n_estimators=200, criterion="entropy")
```

```python
randomforest.fit(X, y)
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = tfidf.transform(test_transform)
predictions = randomforest.predict(test_dataset)
```

```python
from sklearn.metrics import accuracy_score
```
